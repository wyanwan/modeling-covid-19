```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("covid-19-los-angeles-county.csv")
df.head()
```

```python
df.columns
```

```python
cumulative = pd.DataFrame(df.loc[df['day'] == 751])
```

```python
df['death_to_case_ratio'] = (df['deaths']/df['cases'])
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Los Angeles County Visualizations
```

```python
#Death to case ratio
df['7_day_dtc_ratio_avg'] = df['death_to_case_ratio'].rolling(7).mean()
fig = sns.lineplot(x=df['date'], y=df['7_day_dtc_ratio_avg'])
fig.set_xlabel("Date", fontsize = 20)
fig.set_ylabel("Death to Case Ratio", fontsize = 20)
fig.set_title("Death to Case Ratio in Los Angeles County", fontsize = 25)
fig.tick_params(labelsize=20)
sns.set(rc = {'figure.figsize':(20,8)})
```

```python
df['new_deaths'] = 0 
for i in range(1,len(df)):
    df['new_deaths'].iloc[i] = df['deaths'].iloc[i] - df['deaths'].iloc[i-1]
    
df['new_cases'] = 0 
for i in range(1,len(df)):
    df['new_cases'].iloc[i] = df['cases'].iloc[i] - df['cases'].iloc[i-1]
```

```python
#New Deaths with 7 day average
df['7_day_death_avg'] = df['new_deaths'].rolling(7).mean()
plt.plot(df['date'], df['7_day_death_avg'])
plt.xlabel("Date", fontsize=20)
plt.ylabel("New Deaths", fontsize=20)
plt.title("New Deaths in Los Angeles County", fontsize=25)
plt.tick_params(labelsize=20)
sns.set(rc = {'figure.figsize':(20,8)})
```

```python
# New cases in 7 day average
df['7_day_case_avg'] = df['new_cases'].rolling(7).mean()
plt.plot(df['date'], df['7_day_case_avg'])
plt.xlabel("Date", fontsize=20)
plt.ylabel("New Cases", fontsize=20)
plt.title("New Cases in Los Angeles County", fontsize=25)
plt.tick_params(labelsize=20)
```

```python
# Cumulative cases
fig1 = sns.lineplot(x=df['date'], y=df['cases'])
fig1.set_xlabel("Date", fontsize = 20)
fig1.set_ylabel("Cumulative Cases", fontsize = 20)
fig1.set_title("Cumulative Cases in Los Angeles County", fontsize = 25)
fig1.tick_params(labelsize=20)
```

```python
#Cumulative deaths
fig1 = sns.lineplot(x=df['date'], y=df['deaths'])
fig1.set_xlabel("Date", fontsize = 20)
fig1.set_ylabel("Cumulative Deaths", fontsize = 20)
fig1.set_title("Cumulative Deaths in Los Angeles County", fontsize = 25)
fig1.tick_params(labelsize=20)
```
