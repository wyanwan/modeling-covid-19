```python
# Importing pandas
import pandas as pd
```

### Updating covid-19-us.csv

```python
# Reading in CSV
us = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us.csv')
```

```python
# Removing entries before 1/26/2020
us = us.drop([0, 1, 2, 3, 4])
```

```python
# Resetting indices
us = us.reset_index(drop=True)
```

```python
# Adding a day column
day_us = range(1, len(us) + 1)
us.insert(1, 'day', day_us)
```

```python
# Exporting to CSV
us.to_csv('covid-19-us.csv', index=False)
```

### Updating covid-19-states.csv

```python
# Reading in CSV
us_states = pd.read_csv('https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-states.csv')
```

```python
# Removing entries before 1/26/2020
us_states = us_states.drop([col for col in range(0,8)])
```

```python
# Resetting indices
us_states = us_states.reset_index(drop=True)
```

```python
# Adding a day column
day_us_states = pd.to_datetime(us_states['date']).sub(pd.Timestamp('2020-01-25')).dt.days
us_states.insert(1, 'day', day_us_states)
```

```python
# Exporting to CSV
us_states.to_csv('covid-19-states.csv', index=False)
```
