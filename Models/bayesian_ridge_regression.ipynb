```python
# Imports
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.linear_model import BayesianRidge
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import cross_val_score
```

#### Data preprocessing

```python
# Reading in LA County case and death data
# Dates 1/26/2020 through 2/22/2022 for this iteration
la = pd.read_csv('https://raw.githubusercontent.com/nmonozon/modeling-covid-19/main/Datasets/covid-19-los-angeles-county.csv', index_col='date')[0:759]
```

```python
la.tail()
```

```python
# Column for new cases per day
la['new_cases'] = 0 
for i in range(1,len(la)):
    la['new_cases'].iloc[i] = la['cases'].iloc[i] - la['cases'].iloc[i-1]
```

```python
la['week'] = la['day'] // 7
weekly_la = la.groupby('week')[['new_cases']].sum()
weekly_la['one_week_before'] = weekly_la['new_cases'].shift(1)
weekly_la['change'] = weekly_la['new_cases'].shift(1) - weekly_la['new_cases'].shift(2)
weekly_la.dropna(inplace = True)
weekly_la = weekly_la.iloc[1::]
```

```python
weekly_la
```

#### Bayesian Ridge Regression model

```python
# Cross validating Bayesian Ridge model
bayesian = linear_model.BayesianRidge()
cv = cross_val_score(bayesian, weekly_la.drop(columns = ['new_cases']), 
                     weekly_la['new_cases'], scoring = 'neg_mean_absolute_error', cv = 5)
cv
```

```python
# Fitting Bayesian Ridge Regression to data
bayesian.fit(weekly_la.drop(columns = ['new_cases']), weekly_la['new_cases'])
preds_bayesian = bayesian.predict(weekly_la.drop(columns = ['new_cases']))
```

```python
# Visualization
plt.figure(figsize = (16,10))
plt.plot(weekly_la['new_cases'], label = 'Confirmed Weekly Cases')
plt.plot(preds_bayesian, label = 'Bayesian Ridge Regression Predicted Cases', color='purple', linestyle='dashed')
plt.xlabel('Weeks Since 1/26/2020', fontsize=18)
plt.ylabel('New Cases', fontsize=18)
plt.title('Bayesian Ridge Regression Weekly COVID-19 Predicted Cases for LA County', fontsize=22)
plt.legend(prop={'size': 14})
# plt.savefig("bayesian_ridge_regression.png")
# files.download("bayesian_ridge_regression.png") 
```

```python
# MAE and MSE
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
print('MAE:', mean_absolute_error(weekly_la['new_cases'], preds_bayesian))
print('MSE:', mean_squared_error(weekly_la['new_cases'], preds_bayesian))
```

```python
# Week 
week = 108
```

```python
# R^2
from sklearn.metrics import r2_score
print('R^2:', r2_score(weekly_la['new_cases'], preds_bayesian))
```

```python
# Actual data from Week 108
week_actual = weekly_la.loc[week]
week_actual
```

```python
# Predicted data for Week 108
bayesian_prediction = bayesian.predict(pd.DataFrame(np.array([[week_actual['new_cases'], 
                      week_actual['new_cases'] - week_actual['one_week_before']]]), columns = weekly_la.columns[1:]))
int(bayesian_prediction)
```
